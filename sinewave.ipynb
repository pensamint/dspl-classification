```python
import numpy as np
import matplotlib.pyplot as plot
import h5py

 

# Get x values of the sine wave
time        = np.arange(0, 10, 0.1);

# Amplitude of the sine wave is sine of a variable like time
amplitude   = np.sin(time)

# Plot a sine wave using time and amplitude obtained for the sine wave
plot.plot(time, amplitude)

# Give a title for the sine wave plot
plot.title('Sine wave')

# Give x axis label for the sine wave plot
plot.xlabel('Time')

# Give y axis label for the sine wave plot
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

# Display the sine wave
plot.show()
```

```python
dataset = np.column_stack((time, amplitude))

# The first step to creating a HDF5 file is to initialise it.
# The first argument provides the filename and location, the second the mode.
# w is for write access.

hf = h5py.File('sine_wave.h5', 'w')
# This creates a file object, hf, which has a bunch of associated methods.
# One is create_dataset which takes the numpy array and creates the dataset

hf.create_dataset('sine_wave', data = dataset)

hf.close()
```
